# Pie Charts
### Visualizing Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating simulated data
nbins = 5
totalN = 100

rawdata = np.ceil(np.logspace(np.log10(1/2), np.log10(nbins -0.01), totalN))

# Prepare the data for Pie Chart
uniquenums = np.unique(rawdata)
data4pie = np.zeros(len(uniquenums))
                  
for i in range(len(uniquenums)):
    data4pie[i] = sum(rawdata == uniquenums[i])
```

```python
plt.pie(data4pie, labels = 100 * data4pie / sum(data4pie))
plt.show()
```

```python
# Another option
plt.pie(data4pie, labels = ['zero', 'one', 'two', 'three', 'four'], explode = [0.02, 0.02, 0.02, 0.02, 0.02])
plt.show()
```

### For Continous Data

```python
# Generate log-normal distribution
data = np.exp(np.random.randn(1000)/10)


# Generate bins using histogram
histout = np.histogram(data, bins = 6)

plt.pie(histout[0])
plt.show()
```
