### BoxPlot Challege

Create two boxplot, one for random gaussian distribution and another for random uniform distribution


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Size of the sample
m = 100
n = 10

# Create the data
data_gauss = np.zeros((m, n)) 
data_unif = np.zeros((m, n))

for i in range(n):
    data_gauss[:, i] = np.random.randn(m)

for i in range(n):
    data_unif[:, i] = np.random.rand(m)
```

```python
# Create the Boxplot using MatPlotLib

fig, ax = plt.subplots(1, 2, figsize=(8, 2));

ax[0].boxplot(data_gauss)
ax[0].set_title("Gaussian BoxPlot")
                       
ax[1].boxplot(data_unif)
ax[1].set_title("Uniform Boxplot")

plt.show()
                       
                       
```
