# Bar Plots

### Visualizing Bar Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
m = 30
n = 6

data = np.zeros((m, n))
```

```python
for i in range(n):
    data[:, i] = m*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10, 4));


# Grafico de Barras (solo barras)
ax[0].bar(range(n), np.mean(data, axis=0))
ax[0].set_title('Bar Plot')


# Gráfico de barras de error
ax[1].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle = '')
ax[1].set_title('Errorbar plot')

ax[2].bar(range(n), np.mean(data, axis=0))
ax[2].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='', color='green')
ax[2].set_title('Errorbar plot + data')
```
