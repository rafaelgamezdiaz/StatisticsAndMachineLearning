# Histograms
### Visualizing Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000

# Generatint the fake data
data = np.exp(np.random.randn(n)/2)
```

Histogram

```python
# numbers bins
k = 40

plt.plot(data, 's')
plt.show
```

```python
plt.hist(data, bins= k)
plt.show
```

```python
plt.hist(data, bins= 10)
plt.show
```

### Another option


```python
y, x = np.histogram(data, bins=50)

# compute the bins centers: if we plot plt.plot(x, y) without to use the next line, we will get an error
# due to the np.histogram requires equals quantity of x and y, but x has 1 more, due to in contains a 
# point by each boundary of the bar. Then in the next line we compute the center of each beans and obtain a new xx
# to use in the histogram plot
xx = (x[1:] + x[:-1])/2

plt.plot(xx, y)
plt.show
```
