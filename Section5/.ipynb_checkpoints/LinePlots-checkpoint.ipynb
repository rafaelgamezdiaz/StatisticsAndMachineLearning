# Line Plots
### Visualizing Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating the data

# Number of data points
n = 100

data1 = np.exp(np.random.randn(n)/2)
data2 = np.exp(np.random.randn(n)/10)
data3 = np.exp(np.random.randn(n)/2 + 1)
```

```python
# Plots of the histograms

# bins size
k = 20

plt.hist(data1, bins = k)
plt.hist(data2, bins = k)
plt.hist(data3, bins = k)

plt.show()
```

```python
# Histograms discretization for datasets
hist1, bins_edges1 = np.histogram(data1, bins = k)
bins_center1 = (bins_edges1[:-1] + bins_edges1[1:]) / 2
hist1 = 100*hist1 / sum(hist1) # Convert to probability

hist2, bins_edges2 = np.histogram(data2, bins = k)
bins_center2 = (bins_edges2[:-1] + bins_edges2[1:]) / 2
hist2 = 100*hist2 / sum(hist2) # Convert to probability

hist3, bins_edges3 = np.histogram(data3, bins = k)
bins_center3 = (bins_edges3[:-1] + bins_edges3[1:]) / 2
hist3 = 100*hist3 / sum(hist3) # Convert to probability

plt.plot(bins_center1, hist1, 's-', label = 'data1')
plt.plot(bins_center2, hist2, 'o-', label = 'data2')
plt.plot(bins_center3, hist3, '^-', label = 'data3')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```
