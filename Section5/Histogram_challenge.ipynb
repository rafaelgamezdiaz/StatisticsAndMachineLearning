```python
import numpy as np
import matplotlib.pyplot as plt

# Number of simulated observations
n = 1000

# Generate the data
data = np.exp(np.random.randn(n)/2)

# Number of bins
k = 40

# We get an initial visualization of the data
plt.plot(data, 's')
plt.show()
```

```python
# one way to visualize the Histogram (using matplotlib)
plt.hist(data, k, density=True) # with density = True we make the Proiportion histogram (to 1)
plt.show()
```

```python
# Another way (using numpy)
# here hist is an array with the counts on each bin, and bin_edges is an array containing each bin edge of the histogram plot 
hist, bin_edges = np.histogram(data, bins = k)

# get the bin width
bin_width = bin_edges[1] - bin_edges[0]

# as the number of bin_edges is the is the number of hist array plus 1
# then we get the center of each bin
bin_centers = (bin_edges[1:] + bin_edges[:-1])/2

# Visualize the histogram using a plot bar
plt.bar(bin_centers, hist, width = bin_width)
plt.title('Histogram Count')
plt.ylabel('count values')
plt.show()
```

Normalizing the data to get the Proportion Percentage Histogram

```python
# Normalizing the values on each bin
hist_norm = (hist * 100) / np.sum(hist)

# Visualizing the histogram
plt.bar(bin_centers, hist_norm, width = bin_width)
plt.title('Proportion Percentage Histogram')
plt.ylabel('Percentage values')
plt.show()
print(np.sum(hist_norm))
```

Normalizing the data to get the Proportion Histogram (to 1)

```python
# Normalizar el histograma
hist_norm = hist / (np.sum(hist) * bin_width)

# Visualizing the histogram
plt.bar(bin_centers, hist_norm, width = bin_width)
plt.title('Proportion Histogram')
plt.ylabel('Proportion Values')
plt.show()
print(np.sum(hist_norm))
```
