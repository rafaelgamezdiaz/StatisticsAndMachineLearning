# Bar Plots

### Visualizing Box and Whisker Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
# Create the data

m = 30 # rows
n = 6 # columns

data = np.zeros((m, n))

for i in range(n):
    data[:, i] = m*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
data.size
```

```python
# Boxplor

plt.boxplot(data)
plt.show()
```

```python
# Plot with Seaborn

sb.boxplot(data = data, orient = 'v')
plt.show()
```

```python
# Or using Pandas and Seaborn
df = pd.DataFrame(data, columns = ["zero", "one", "two", "three", "four", "five"])
sb.boxplot(data = df, orient='h')
plt.show()
```
