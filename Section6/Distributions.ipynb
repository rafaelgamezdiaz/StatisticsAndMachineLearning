### Analityc Gaussian (normal) Distribution

```python
# Gaussian

# number of discretizations
N = 1001
x = np.linspace(-4, 4, N)
gausdist = stats.norm.pdf(x)

plt.plot(x, gausdist)
plt.title("Analityc Gaussian (normal) Distribution")
plt.show()
```

Is this a probability distribution?

```python
print(sum(gausdist))
```

The sum is not 1.

To get the area under the curve we can sum the height on each bar (imagine the curve as a sum on many vertical bars), multipliyed by a bins width. Here we are considerating many rectangles over the distribution graph.

As the bin size is common on each sumatory term, we can extract it as a common value. Then we can get the area multiplying the sum of all gausdist termn with by the bin size.

```python
# First we need to get the diference between two points in the x data
bin_size = np.diff(x[:2])# The sum is not 1. we neet to normalize

area = sum(gausdist) * bin_size
print(area)
```

### Empirical Normal Distribution

```python
# Normally distributed random numbers

#parameters
stretch = 1 # variance (square of standard deviation)
shift = 5 # mean
n = 1000

# create the data
data = stretch * np.random.randn(n) + shift

# plot the data
plt.hist(data, bins = 50)
plt.title("Empirical Normal Distribution")
plt.show()
```

```python
print(np.mean(data))
```

```python
## Uniformerly - distributed numbers

# parameters
stretch = 2 # not the variance
shift = .5
n = 10000

# create data
data = stretch * np.random.rand(n) + shift - stretch/2


#plot data
fig, ax = plt.subplots(2, 1, figsize = (5,6))

ax[0].plot(data, '.', markersize=1)
ax[0].set_title('Uniform Data Values')

ax[1].hist(data, 25)
ax[1].set_title('Uniform Data Histogram')

plt.show()
```

### Log - Normal Distribution

```python
## Log-normal distribution

N = 1001
x = np.linspace(0,10,N)
lognormdist = stats.lognorm.pdf(x, 1)

plt.plot(x, lognormdist)
plt.title('Analitic Log-Normal Distribution')
plt.show()
```

```python
# Empirical log-romal distribution

shift = 5 # the mean
stretch = 0.5
n = 2000

# Generate data
# To generate a log-normal fort we generate a gaussian random number distribution, 
# and then aply the exponential function to it

data = np.random.randn(n) + shift
data = stretch * np.exp(data)

#data = np.random.randn(n)
#data = stretch * np.exp(data) + shift

# plot data
fig, ax = plt.subplots(2, 1, figsize = (4, 6))
ax[0].plot(data, '.')
ax[0].set_title('Log-Normal Data Values')

ax[1].hist(data, bins = 25)
ax[1].set_title('LogNormal Data Histogram')

plt.show()
```

### Binomial Distribution


A binomial distribution is the probability of K heads in N coin tooses 
given a probability of p heads (e.g. 0.5 is a fair coin)

```python
n = 10 # number of coin tooses
p = 0.5 # probability of p heads

x = range(n + 2)
bindist = stats.binom.pmf(x, n, p)

plt.bar(x, bindist)
plt.title('Binomial Distribution (n=%s, p=%g)'%(n,p))
plt.show()
```

### t Distribution

```python
x = np.linspace(-4, 4, 1001)
df = 200
t = stats.t.pdf(x, df)

#df2 = 1
#t2 = stats.t.pdf(x, df2)

plt.plot(x, t, '.', markersize=2, color='blue')
plt.xlabel('t-value')
plt.ylabel('P(t | H$_0$)')
plt.title('t(%g) distribution'%df)
plt.show()
```

```python
bin_size = np.diff(x[:2])
t_norm = t * bin_size
print(sum(t_norm))

plt.plot(x, t_norm, '.', markersize=2, color='red')
plt.xlabel('t-value')
plt.ylabel('P(t | H$_0$)')
plt.title('t(%g) distribution'%df)
plt.show()
```

### F Distribution

```python
# Parameters

num_df = 4 # number of degree of freedom
den_df = 100 # denominator df

# values to evaluate
x = np.linspace(0, 10, 10001)

# F distribution
fdist = stats.f.pdf(x, num_df, den_df)

plt.plot(x, fdist)
plt.title(f'F({num_df}, {den_df}) distribution')
plt.xlabel('F value')
plt.show()
```
