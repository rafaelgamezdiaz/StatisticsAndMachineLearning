### Analityc Gaussian (normal) Distribution

```python
# Gaussian

# number of discretizations
N = 1001
x = np.linspace(-4, 4, N)
gausdist = stats.norm.pdf(x)

plt.plot(x, gausdist)
plt.title("Analityc Gaussian (normal) Distribution")
plt.show()
```

Is this a probability distribution?

```python
print(sum(gausdist))
```

The sum is not 1.

To get the area under the curve we can sum the height on each bar (imagine the curve as a sum on many vertical bars), multipliyed by a bins width. Here we are considerating many rectangles over the distribution graph.

As the bin size is common on each sumatory term, we can extract it as a common value. Then we can get the area multiplying the sum of all gausdist termn with by the bin size.

```python
# First we need to get the diference between two points in the x data
bin_size = np.diff(x[:2])# The sum is not 1. we neet to normalize

area = sum(gausdist) * bin_size
print(area)
```

### Empirical Normal Distribution

```python
# Normally distributed random numbers

#parameters
stretch = 1 # variance (square of standard deviation)
shift = 5 # mean
n = 1000

# create the data
data = stretch * np.random.randn(n) + shift

# plot the data
plt.hist(data, bins = 25)
plt.title("Empirical Normal Distribution")
plt.show()
```

```python
print(np.mean(data))
```

```python
## Uniformerly - distributed numbers

# parameters
stretch = 2 # not the variance
shift = .5
n = 10000

# create data
data = stretch * np.random.rand(n) + (shift - stretch)/2


#plot data
fig, ax = plt.subplots(2, 1, figsize = (5,6))

ax[0].plot(data, '.', markersize=1)
ax[0].set_title('Uniform Data Values')

ax[1].hist(data, 25)
ax[1].set_title('Uniform Data Histogram')

plt.show()
```

```python
## Log-normal distribution

N = 1001
x = np.linspace(0,10,N)
lognormdist = stats.lognorm.pdf(x, 1)

plt.plot(x, lognormdist)
plt.title('Analitic Log-Normal Distribution')
plt.show()
```
