# Central Tendency (Unsupervise Learning)

### Unsupervise Learning Challenge!

Compare mean and median for distributions with:
1) 'small' outliers
2) 'large' outliers
3) small-N datasets
4) large-N datasets

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def dataHist(data, nbins):
    y, x = np.histogram(data, nbins)
    x = (x[1:] + x[:-1]) / 2
    return x, y
```

```python
def get_statistics(dataset):
    mode_value, mode_count = stats.mode(dataset, axis=None, keepdims=False)
    return np.mean(dataset), np.median(dataset), mode_value
```

```python
def show_plots(x, y, mean_value, median_value, mod_value, i, j, title):
    ax[i][j].plot(x, y, 'b')
    ax[i][j].plot([mean_value, mean_value], [0, max(y)], 'b--')
    ax[i][j].plot([median_value, median_value], [0, max(y)], 'r--')
    ax[i][j].plot([mod_value, mod_value], [0, max(y)], 'k--')
    ax[i][j].set_title(title)
    ax[i][j].set_xlim(0,15)
```

```python
def log_normal(n, stretch, shift):
    return np.exp(stretch * np.random.randn(n) + shift)
```

```python
# Using a log-normal distribution

# parameters
stretch = 0.7
shift = 0
nbins = 30
small_outlier = 25
large_outlier = 1000
```

```python
# Graphs
fig, ax = plt.subplots(2, 2, figsize=(12,12))

# TESTS LARGE DATASET
# Adding 'smal' ourliers
n = 1000
dataset1 = log_normal(n, stretch, shift)
dataset1 = np.append(dataset1, small_outlier)
x1, y1 = dataHist(dataset1, nbins)
mean1, median1, mod1 = get_statistics(dataset1)
show_plots(x1, y1, mean1, median1, mod1, 0, 0, 'Large N, Small Outliers')


# TESTS LARGE DATASET
# Adding 'large' ourliers
n = 1000
dataset2 = log_normal(n, stretch, shift)
dataset2 = np.append(dataset2, large_outlier) # small_outlier large_outlier
x2, y2 = dataHist(dataset2, nbins)
mean2, median2, mod2 = get_statistics(dataset2)
show_plots(x2, y2, mean2, median2, mod2, 0, 1, 'Large N, Large Outliers')


# TESTS SMALL DATASET
# Adding 'small' ourliers
n = 40
dataset3 = log_normal(n, stretch, shift)
dataset3 = np.append(dataset3, small_outlier)
x3, y3 = dataHist(dataset3, nbins)
mean3, median3, mod3 = get_statistics(dataset3)
show_plots(x3, y3, mean3, median3, mod3, 1, 0, 'Small N, Small Outliers')

# TESTS SMALL DATASET
# Adding 'large' ourliers
n = 40
dataset4 = log_normal(n, stretch, shift)
dataset4 = np.append(dataset4, large_outlier)
x4, y4 = dataHist(dataset4, nbins)
mean4, median4, mod4 = get_statistics(dataset4)
show_plots(x4, y4, mean4, median4, mod4, 1, 1, 'Small N, Large Outliers')

plt.show()
```
