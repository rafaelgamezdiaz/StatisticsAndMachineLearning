# Computing Dispersion

### Section6: Descriptive Statistics

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def dataHist(data, nbins):
    y, x = np.histogram(data, nbins)
    x = (x[1:] + x[:-1]) / 2
    return x, y
```

```python
# Create some data distribution
N = 10001 # number of data points
nbins = 30 # number of histogram builds

d1 = np.random.randn(N) - 1
d2 = 3 * np.random.randn(N)
d3 = np.random.randn(N) + 1

# Histograms
x1, y1 = dataHist(d1, nbins)
x2, y2 = dataHist(d2, nbins)
x3, y3 = dataHist(d3, nbins)

# Ploting
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'r')
plt.plot(x3, y3, 'k')

plt.xlabel('Data Values')
plt.ylabel('Data Counts')
plt.show()
```

```python
# Other way

meanval = 10.2
stdval = 7.5
numsamp = 123

# this
np.random.normal(meanval, stdval, numsamp)

# is equivalent to
np.random.randn(numsamp) * stdval + meanval
```

### Calculating the Mean

```python
# Calculating the Mean
mean_d1 = sum(d1) / len(d1)
mean_d2 =  np.mean(d2)
mean_d3 = np.mean(d3)

# Plot them
plt.plot(x1, y1, 'b', x2, y2, 'r', x3, y3, 'k')
plt.plot([mean_d1, mean_d1], [0, max(y1)], 'b--')
plt.plot([mean_d2, mean_d2], [0, max(y2)], 'r--')
plt.plot([mean_d3, mean_d3], [0, max(y3)], 'k--')

plt.xlabel('Data Values')
plt.ylabel('Data Counts')
plt.show()
```

### Standard Deviation

```python
stds = np.zeros(3)

stds[0] = np.std(d1, ddof=1) # it is important to define the dd0f = 1, to calculate correctly the std for the sample
stds[1] = np.std(d2, ddof=1)
stds[2] = np.std(d3, ddof=1)

plt.plot(x1, y1, 'b', x2, y2, 'r', x3, y3, 'k')
plt.plot([mean_d1, mean_d1], [0, max(y1)], 'b--')
plt.plot([mean_d2, mean_d2], [0, max(y2)], 'r--')
plt.plot([mean_d3, mean_d3], [0, max(y3)], 'k--')

# now add the stds
plt.plot([mean_d1 - stds[0], mean_d1 + stds[0]], [0.45*max(y1), 0.45*max(y1)], 'b', linewidth=4)
plt.plot([mean_d2 - stds[1], mean_d2 + stds[1]], [0.5*max(y1), 0.5*max(y1)], 'r', linewidth=4)
plt.plot([mean_d3 - stds[2], mean_d3 + stds[2]], [0.55*max(y1), 0.55*max(y1)], 'k', linewidth=4)

plt.xlabel('Data Values')
plt.ylabel('Data Counts')
plt.show()
```
